```python
import requests
from bs4 import BeautifulSoup
import csv

def get_facility(soup, facility_name):
    try:
        facilities_section = soup.find("div", class_="bui-u-inline")
        facility = facilities_section.find('span', text=facility_name)
        if facility:
            return 'Yes'
        else:
            return 'No'
    except AttributeError:
        return 'No'

def get_breakfast_cuisine(soup, cuisine_number):
    try:
        breakfast_section = soup.find("div", class_="bui-grid bui-grid--small-12 bui-grid--xsmall-1 bui-spacer--large-2 bui-u-pt-16")
        cuisine = breakfast_section.find_all("li", class_="bui-list__item")
        if cuisine_number <= len(cuisine):
            return cuisine[cuisine_number - 1].text
        else:
            return ''
    except AttributeError:
        return ''

# Define the cities and their URLs
cities = {
    "Dubai, United Arab Emirates": "https://www.booking.com/city/ae/dubai.html",
    "London, United Kingdom": "https://www.booking.com/city/gb/london.html",
    "Kuala Lumpur, Malaysia": "https://www.booking.com/city/my/kuala-lumpur.html",
    "Manchester": "https://www.booking.com/city/gb/manchester.html",
    "New Delhi": "https://www.booking.com/city/in/new-delhi.html",
    "Birmingham": "https://www.booking.com/city/gb/birmingham.html",
    "Berlin": "https://www.booking.com/city/de/berlin.html",
    "Sydney": "https://www.booking.com/city/au/sydney.html",
    "Melbourne": "https://www.booking.com/city/au/melbourne.html",
    "Paris": "https://www.booking.com/city/fr/paris.html",
    "Tokyo": "https://www.booking.com/city/jp/tokyo.html",
    "Toronto": "https://www.booking.com/city/ca/toronto.html"
}

with open('Data2.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['sno', 'Hotel name', 'Free wifi', 'Family rooms', 'Non Smoking Rooms', 'Restaurant', 'Bar', 'Heating', 'Lift', 'Breakfast Cuisine1', 'Breakfast Cuisine2', 'Breakfast Cuisine3'])

    sno = 1
    for city, url in cities.items():
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extracting data from the webpage
        hotel_name = soup.find("h2", class_="hp__hotel-name")
        free_wifi = get_facility(soup, "Free WiFi")
        family_rooms = get_facility(soup, "Family rooms")
        non_smoking_rooms = get_facility(soup, "Non-smoking rooms")
        restaurant = get_facility(soup, "Restaurant")
        bar = get_facility(soup, "Bar")
        heating = get_facility(soup, "Heating")
        lift = get_facility(soup, "Lift")
        breakfast_cuisine1 = get_breakfast_cuisine(soup, 1)
        breakfast_cuisine2 = get_breakfast_cuisine(soup, 2)
        breakfast_cuisine3 = get_breakfast_cuisine(soup, 3)

        # Writing data to CSV
        writer.writerow([sno, hotel_name, free_wifi, family_rooms, non_smoking_rooms, restaurant, bar, heating, lift, breakfast_cuisine1, breakfast_cuisine2, breakfast_cuisine3])
        sno += 1

print("Data has been written to Data2.csv")
```
